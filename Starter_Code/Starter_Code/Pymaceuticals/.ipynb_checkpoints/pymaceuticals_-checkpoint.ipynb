```python
#Add commands so plots generated by matplotlib show here
#%matplotlib inline 
#%matplotlib notebook
#%matplotlib widget
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data of mouse data and study results into a single DataFrame,we wannt to display number of unique mice IDs.
combine_mouse_study_df = pd.merge(mouse_metadata, study_results, how='outer', on="Mouse ID")

# Display the data table for preview
combine_mouse_study_df.head()
```

```python
#Remove duplicates of any mice by the ID number.
# Our data should be uniquely identified by Mouse ID and Timepoint
dupe_mice = combine_mouse_study_df.loc[combine_mouse_study_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID'].unique()
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupe_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# Find the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen. 
# So we need to use Tutor Volume (mm3) and Drug Regimen in our code to find mean,median, varience, std, and SEM.  
mean = clean_data_df['Tumor Volume (mm3)'].groupby(clean_data_df['Drug Regimen']).mean()
median = clean_data_df['Tumor Volume (mm3)'].groupby(clean_data_df['Drug Regimen']).median()
var = clean_data_df['Tumor Volume (mm3)'].groupby(clean_data_df['Drug Regimen']).var()
std = clean_data_df['Tumor Volume (mm3)'].groupby(clean_data_df['Drug Regimen']).std()
sem = clean_data_df['Tumor Volume (mm3)'].groupby(clean_data_df['Drug Regimen']).sem()

#Alright make the summary Dataframe and make your labels

summary_stats = pd.DataFrame({"Mean Tumor Vol.":mean,
                              "Median Tumor Vol.":median,
                              "Varience Tumor Vol.":var,
                              "Std. Deviation Tumor Vol.":std,
                              "Std. Error of Mean Tumor Vol.":sem})
#Display the summary statistics in a single summary Dataframe
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
#Use the clean_data_df from summary statitics to use in this new summary 
agg_drug_sum_df = clean_data_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean","median","var","std","sem"])
agg_drug_sum_df
```

## Quartiles, Outliers and Boxplots

```python
#start by getting the last(greatest) timepoint for each mouse
max_tumor = clean_data_df.groupby(["Mouse ID"])['Timepoint'].max()
max_tumor = max_tumor.reset_index()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_tumor = clean_data_df.groupby(["Mouse ID"])['Timepoint'].max()
max_tumor = max_tumor.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = max_tumor.merge(clean_data_df,on=['Mouse ID','Timepoint'],how="left")

merged_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)

drug_treatments = ["Ramicane", "Capomulin", "Ceftamin", "Infubinol"]

# Create empty list to fill with tumor vol data (for plotting)

volume_lists = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
#Use the quantile() method and Interquartile Range (IQR) method



# Locate the rows which contain mice on each drug and get the tumor volumes

#merged_data.loc[merged_data["Drug Regimen"] == "Ramicane"]
for drug in drug_treatments:
    temptdf =  merged_data.loc[merged_data["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    volume_lists.append(temptdf)   
    #Use the quantile() method and Interquartile Range (IQR) method
    quartiles = temptdf.quantile([.25,.50,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    IQR = upperq - lowerq
    # Determine outliers using upper and lower bounds

    low_bound = lowerq - (1.5*IQR)
    upper_bound = upperq + (1.5*IQR)
   

    # add subset 
    outlier_occupancy = temptdf.loc[(temptdf< low_bound) | (temptdf > upper_bound)]
    print(f"{drug} potential outliers: {outlier_occupancy}")
# Determine outliers using upper and lower bounds

```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(volume_lists, labels= drug_treatments)
plt.title("Tumor Volume (mm3) of each mouse")
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
#Gonna use same mouse l509 so we know what we are looking for and make specific data for that mouse
line_l509 = clean_data_df.loc[clean_data_df["Drug Regimen"] == "Capomulin"].loc[clean_data_df["Mouse ID"] == "l509"]

x_axis = line_l509["Timepoint"]
y_axis = line_l509["Tumor Volume (mm3)"]

plt.title("Capomulin treatment for mouse l509")
plt.plot(x_axis, y_axis, linewidth=3)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
#so make a data just for the mice that were treated with Capomulin
capomulin_mice = clean_data_df.loc[clean_data_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID").mean()

scatter_xaxis = capomulin_mice["Weight (g)"]
scatter_yaxis = capomulin_mice["Tumor Volume (mm3)"]

plt.scatter(scatter_xaxis, scatter_yaxis, color = "black")
plt.title("Capomulin mouse weight vs. Total Volume (mm3)")
plt.xlabel("weight (g)")
plt.ylabel("Average Tumor Vol (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
#So lets make the scatter plot... again
capomulin_mice = clean_data_df.loc[clean_data_df["Drug Regimen"] == "Capomulin"].groupby("Mouse ID").mean()

scatter_xaxis = capomulin_mice["Weight (g)"]
scatter_yaxis = capomulin_mice["Tumor Volume (mm3)"]

#make the regression line
#obtain m (slope) and b (intercept) of linear regression then plot
m, b = np.polyfit(scatter_xaxis, scatter_yaxis, 1)
plt.plot(scatter_xaxis, m*scatter_xaxis+b, color="red")

plt.scatter(scatter_xaxis, scatter_yaxis, color = "blue")
plt.title("Capomulin mouse weight vs. Total Volume (mm3)")
plt.xlabel("weight (g)")
plt.ylabel("Average Tumor Vol (mm3)")


#Display Scatter plot
plt.show()
```

```python
#find the correlation mouse weight and the average tumor volume (mm3).

corr_mouse_weight_vol = round(st.pearsonr(scatter_xaxis,scatter_yaxis)[0],2)
#corr_version_2
print(f"The correlation between mouse weight and the average tumor volume is {corr_mouse_weight_vol}")
```
